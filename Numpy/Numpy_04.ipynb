```python
x = numpy.linspace(0,10,200)
y = numpy.sin(x)
```

```python
plt.figure(figsize = (6,2))
plt.plot(x, y, "--r")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Sine Graph")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (6,2))
plt.plot(x, y, "--b")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Sine Graph")
plt.grid()
plt.show()
```
