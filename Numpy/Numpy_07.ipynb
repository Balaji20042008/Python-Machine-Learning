```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers

output_file("test.html")

p = figure(title = "test_flower")
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.circle(x = flowers["petal_length"], y = flowers["petal_width"])
show(p)
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

x = [1, 2, 3, 4, 5]
y = [9, 13, 17, 21, 25]

from bokeh.plotting import figure, output_file, show

output_file("line.html")

p = figure(title = "line graph")
p.xaxis.axis_label = "x-axis"
p.yaxis.axis_label = "y-axis"
p.scatter(x = x, y = y, fill_color = "red", legend_label = "red points", size = 10)
show(p)
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

# x = [1, 2, 3, 4, 5]
# y = [9, 13, 17, 21, 25]

from bokeh.plotting import figure, output_file, show

import seaborn as sns

tips = sns.load_dataset("tips")

output_file("tips.html")

p = figure(title = "tips graph")
p.xaxis.axis_label = "x-axis"
p.yaxis.axis_label = "y-axis"
p.scatter(x = tips.total_bill, y = tips.tip,  size = 3 * tips["size"])
show(p)
```
