## # EDA - Student Performance Dataset

### 1. Problem Statement

- This project understands how the student's performance (test scores) is affected by other variables such as Gender, Ethnicity, Parental Level of Education, Lunch, Test Preparation Course

### 2. Data Collection

- Dataset Source - https://www.kaggle.com/datasets/spscientist/students-performance-in-exams?datasetId=74977
- The data consists of 8 columns and 100 rows

### 3. Dataset Information

- Gender: Sex of Student -> (Male/Female)
- Race/ethinicity: Ethnicity of Students -> (Group A, B, C, D, E)
- Parental Level of Education: Parent's final education -> (bachelor's degree, some college, master's degree, associate's degree, high school)
- Lunch: Having lunvh before test (Standard or Free/Reduced)
- Test Preparation Course: Complete or Not Complete
- Math Score
- Reading Score
- Writing Score

```python
df["total score"] = df["math score"] + df["reading score"] + df["writing score"]
df["average"] = df["total score"] / 3
```

### # Explore Data Visualization

```python
fig,axis = plt.subplots(1, 2, figsize = (14, 7)) 
plt.subplot(1, 2, 1)
sns.histplot(data = df, x = df["average"], bins = 30, kde = True, color = "green")
plt.subplot(1, 2, 2)
sns.histplot(data = df, x = df["average"], bins = 30, kde = True, hue = "gender")
```

#### Insights/Obseravtions: 
- Female students tend to perform well than male students in the exam.

```python
fig,axis = plt.subplots(1, 3, figsize = (15, 5))
plt.subplot(1, 3, 1)
sns.histplot(data = df, x = df["average"], bins = 30, kde = True, hue = "lunch")
plt.subplot(1, 3, 2)
sns.histplot(data = df[df.gender == "male"], x = df["average"], bins = 30, kde = True, hue = "lunch")
plt.subplot(1, 3, 3)
sns.histplot(data = df[df.gender == "female"], x = df["average"], bins = 30, kde = True, hue = "lunch")
```

#### Insights/Observations:

- Standard Lunch help students perform well in exams
- Standard lunch helps perform well in exams be it a male of female

```python
fig,axis = plt.subplots(1, 3, figsize = (15, 5))
plt.subplot(1, 3, 1)
sns.histplot(data = df, x = df["average"], bins = 30, kde = True, hue = "parental level of education")
plt.subplot(1, 3, 2)
sns.histplot(data = df[df.gender == "male"], x = df["average"], bins = 30, kde = True, hue = "parental level of education")
plt.subplot(1, 3, 3)
sns.histplot(data = df[df.gender == "female"], x = df["average"], bins = 30, kde = True, hue = "parental level of education")
```

####  Insights/Obseravtions:
- In general parent's education don't help student perform well in exam.
- In 2nd plot, that parent's whose education is of associate's degree or master's degree their male child tend to perform well in exam.
- In 3rd plot, that parent's whose education is of bachelor's degree of some high school their female child tend to perform slightly well than others in exam.

```python
fig,axis = plt.subplots(1, 3, figsize = (15, 5))
plt.subplot(1, 3, 1)
sns.histplot(data = df, x = df["average"], bins = 30, kde = True, hue = "race/ethnicity")
plt.subplot(1, 3, 2)
sns.histplot(data = df[df.gender == "male"], x = df["average"], bins = 30, kde = True, hue = "race/ethnicity")
plt.subplot(1, 3, 3)
sns.histplot(data = df[df.gender == "female"], x = df["average"], bins = 30, kde = True, hue = "race/ethnicity")
```

####  Insights/Observations:

- Students of group A and group B tends to perform poorly in exam.
- Students of group A and group B tends to perform poorly in exam irrespective of whether they are male or female.
