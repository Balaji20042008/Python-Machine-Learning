```python
class Perceptron:
    
    def __init__ (self, eta, epochs):
        
        self.weights = np.random.randn(2) * 1e-4
        print("self weights: {}" .format(self.weights))
        self.eta = eta
        self.epochs = epochs
    
    def activation_function (self, inputs, weights):
        
        z = np.dot(inputs, weights)
        return np.where(z > 1, 1, 0)
    
    def fit (self, x, y):
        
        self.x = x
        self.y = y
        
        # x_with_bias = np.c_[self.x, -np.ones((len(self.x), 1))]
        # print("x_with_bias: \n{}" .format(x_with_bias))
        
        for epoch in range(self.epochs):
            print("for epoch: {}" .format(epoch))
            y_hat = self.activation_function(self.x, self.weights)
            print("predicted value: \n{}" .format(y_hat))
            error = y_hat - self.y
            print("error: \n{}" .format(error))
            self.weights = self.weights - self.eta * np.dot(self.x.T, error)
            print("updated weights: \n{}" .format(self.weights))
            
    def predict (self, x):
        
        return self.activation_function(x, self.weights)
```

```python
data = {"x1": [0,0,1,1], "x2": [0,1,0,1], "y": [0,0,0,1]}

AND = pd.DataFrame(data)
AND
```
