```python
def assign_grade(x) :
    
    u = numpy.mean(df_01["Total"])
    v = numpy.std(df_01["Total"])
    
    if x > u + 1.5 * v :
        return 'S'
    elif x <= u + 1.5 * v and x > u + 1.0 * v :
        return 'A'
    elif x <= u + 1.0 * v and x > u + 0.5 * v :
        return 'B'
    elif x <= u + 0.5 * v and x > u - 0.5 * v :
        return 'C'
    elif x <= u - 0.5 * v and x > u - 1.0 * v :
        return 'D'
    elif x <= u - 1.0 * v and x > 40 :
        return 'E'
    else :
        return 'F'
```

```python
df_01["Grade"] = df_01["Total"].apply(assign_grade)
```
