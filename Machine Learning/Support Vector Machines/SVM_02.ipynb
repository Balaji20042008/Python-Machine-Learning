## # Machine Learning - Support Vector Regressor

### Step-01: Import all the libraries that are required to perform Support Vector Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

### Step-02: Import make_regression from sklearn.datasets

```python
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_samples = 1000, n_features = 2, n_targets = 1)
```

```python
x
```

```python
y
```

```python
sns.scatterplot(x = pd.DataFrame(x)[0], y = pd.DataFrame(x)[1], hue = y)
```

### Step-03: Import train_test_split from sklearn.model_selection

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
print("x_train: \n\n {} \n" . format(x_train))
print("x_train.shape: {}" .format(x_train.shape))
```

```python
print("x_test: \n\n {} \n" .format(x_test))
print("x_test.shape: {}" .format(x_test.shape))
```

```python
print("y_train: \n\n {} \n" .format(y_train))
print("y_train.shape: {}" .format(y_train.shape))
```

```python
print("y_test: \n\n {} \n" .format(y_test))
print("y_test.shape: {}" .format(y_test.shape))
```

### Step-04: Import SVR from sklearn.svm

```python
from sklearn.svm import SVR
```

```python
support_vector_regressor = SVR(kernel = "linear")
```

```python
support_vector_regressor.fit(x_train, y_train)
```

```python
y_pred = support_vector_regressor.predict(x_test)
```

```python
y_pred
```

### Step-05: Import r2_score from sklearn.metrics

```python
from sklearn.metrics import r2_score
```

```python
print("R Squared Score: {}" .format(r2_score(y_test, y_pred)))
```

### Step-06: Import GridSearchCV from sklearn.model_selection

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {
    
    "C": [0.1, 1, 10, 100, 100],
    "gamma": [1, 0.1, 0.01, 0.001, 0.0001],
    "epsilon": [0.1, 0.2, 0.3],
    "kernel": ["linear"]
}
```

```python
grid_search_cv = GridSearchCV(support_vector_regressor, param_grid = parameters, scoring = "neg_mean_squared_error", cv = 5, verbose = 3)
```

```python
grid_search_cv.fit(x_train, y_train)
```

```python
grid_search_cv.best_params_
```

```python
y_pred = grid_search_cv.predict(x_test)
```

```python
y_pred
```

```python
print("R Squared Score: {}" .format(r2_score(y_test, y_pred)))
```
