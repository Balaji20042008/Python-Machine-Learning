## # Machine Learning - Interpolation

Data Interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are varuious libraries available that can be used for data interpolation, such as NumPy, SciPy and Pandas. Here is an example of how to perform data interpolation using these libraries:

### 1. Linear Interpolation - NumPy

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 10, 15, 20, 25])
```

```python
plt.scatter(x,y)
```

```python
new_x = np.linspace(1,5,20)
interp_y = np.interp(new_x,x,y)
```

```python
print("new_x: " + str(new_x))
print("interp_y: " + str(interp_y))
```

```python
plt.scatter(new_x, interp_y)
```

### 2. Cubic Interpolation - SciPy

```python
from scipy.interpolate import interp1d
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 8, 27, 64, 125])
```

```python
plt.scatter(x,y)
```

```python
function = interp1d(x, y, kind = "cubic")
```

```python
new_x = np.linspace(1, 5, 20)
interp_y = function(new_x)
```

```python
print("new_x: " + str(new_x))
print("interp_y: " + str(interp_y))
```

```python
plt.scatter(new_x, interp_y)
```

### 3. Polynomial Interpolation - NumPy

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
```

```python
plt.scatter(x,y)
```

```python
function = np.polyfit(x, y, 2)
```

```python
new_x = np.linspace(1, 5, 20)
interp_y = np.polyval(function, new_x)
```

```python
plt.scatter(new_x, interp_y)
```
