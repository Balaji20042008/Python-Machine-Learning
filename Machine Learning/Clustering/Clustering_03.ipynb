## # Machine Learning - DBScan Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings 
warnings.filterwarnings("ignore")
```

### Step-01: Import make_moons from sklearn.datasets

```python
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples = 250, noise = 0.05, random_state = 100)
```

```python
print("x: \n\n {} \n" .format(x))
print("x.shape: {} \n" .format(x.shape))
```

```python
print("y: \n\n {} \n" .format(y))
print("y.shape: {} \n" .format(y.shape))
```

```python
sns.scatterplot(x = x[:,0], y = x[:,1])
plt.xlabel("Feature_01")
plt.ylabel("Feature_02")
plt.title("Feature_01 v/s Feature_02")
```

### Step-02: Import StandardScaler from sklearn.preprocessing

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard_scaler = StandardScaler()
```

```python
x = standard_scaler.fit_transform(x)
```

### Step-03: Import DBSCAN from sklearn.cluster

```python
from sklearn.cluster import DBSCAN
```

```python
db_scan = DBSCAN()
```

```python
db_scan.fit(x)
```

```python
x_labels = db_scan.labels_
```

```python
x_labels
```

```python
sns.scatterplot(x = x[:,0], y = x[:,1], hue = x_labels)
plt.xlabel("Feature_01")
plt.ylabel("Feature_02")
plt.title("Feature_01 v/s Feature_02")
```

```python
sns.scatterplot(x = x[:,0], y = x[:,1], hue = y)
plt.xlabel("Feature_01")
plt.ylabel("Feature_02")
plt.title("Feature_01 v/s Feature_02")
```
