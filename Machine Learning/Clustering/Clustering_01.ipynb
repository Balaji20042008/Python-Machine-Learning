### Step-01: Import all the libraries to perform K Means Clustering

```python
from sklearn.datasets import make_blobs
```

### Step-04: Import KMeans from sklearn.cluster

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for k in range(1,11):
    k_means = KMeans(n_clusters = k, init = "k-means++")
    k_means.fit_transform(x_train)
    wcss.append(k_means.inertia_)
    
for i in range(0,10):
    print("for K = {}, wcss = {}" .format(i+1,wcss[i]))
```

```python
plt.figure(figsize = (6,3))
plt.plot(range(1,11), wcss, "r")
plt.xlabel("K")
plt.ylabel("WCSS")
plt.title("Elbow Curve")
plt.show()
```

```python
k_means = KMeans(n_clusters = 3, init = "k-means++")
```

```python
k_means.inertia_
```

```python
k_means.labels_
```

```python
y_train_labels = k_means.fit_predict(x_train)
y_test_labels = k_means.predict(x_test)
```

```python
plt.figure(figsize = (7,5))
sns.scatterplot(x = x_train[:,0], y = x_train[:,1], hue = y_train_labels)
plt.xlabel("Feature_01")
plt.ylabel("Feature_02")
plt.title("Feature_01 v/s Feature_02")
plt.show()
```

```python
plt.figure(figsize = (7,5))
sns.scatterplot(x = x_test[:,0], y = x_test[:,1], hue = y_test_labels)
plt.xlabel("Feature_01")
plt.ylabel("Feature_02")
plt.title("Feature_01 v/s Feature_02")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
accuracy = []
for k in range(2,11):
    k_means = KMeans(n_clusters = k, init = "k-means++")
    k_means.fit(x_train)
    score = silhouette_score(x_train, k_means.labels_)
    accuracy.append(score)
for k in range(2,11):
    print("for k = {}, silhouette_score = {}" .format(k, accuracy[k-2]))
```

```python
plt.figure(figsize = (7, 5))
sns.lineplot(x = range(2,11), y = accuracy)
plt.xlabel("K")
plt.ylabel("Silhouette Score")
plt.title("K v/s Silhouette Score")
plt.show()
```
